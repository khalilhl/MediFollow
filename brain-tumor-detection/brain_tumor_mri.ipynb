```python
import os
import subprocess
import sys
from pathlib import Path

# Dataset Kaggle : dossier contenant les sous-dossiers yes/ et no/
# Peut être le dossier parent : yes/no seront trouvés dans brain_tumor_dataset/ si besoin
DATA_ROOT = Path(os.environ.get("BRAIN_MRI_DATA_ROOT", r"C:\Users\Khalil\Downloads\dataset"))

subprocess.check_call(
    [sys.executable, "train_brain_tumor.py", "--data_root", str(DATA_ROOT.resolve())],
    cwd=Path.cwd(),
)
```
